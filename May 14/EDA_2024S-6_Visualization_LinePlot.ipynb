# Data Visualization using Python

## Basic plotting in matplotlib

```python
from matplotlib import pyplot as plt
# import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import datetime
# %matplotlib inline
```

In Matplotlib, the figure (an instance of the class plt.Figure) can be thought of as a single container that contains all the objects representing axes, graphics, text, and labels. The axes (an instance of the class plt.Axes) is what we see above: a bounding box with ticks and labels, which will eventually contain the plot elements that make up our visualization.

For all Matplotlib plots, we start by creating a figure and an axes. In their simplest form, a figure and axes can be created as follows:

We'll commonly use the variable name `fig` to refer to a figure instance, and `ax` to refer to an axes instance or group of axes instances.

```python
# fig = plt.figure(figsize=(20,20))
# ax = plt.axes()

fig, ax = plt.subplots()
fig.set_size_inches(30,30)
```

Comparison between `ax` and `plt` in Matplotlib:

|    | ax                                    | plt                                      |
|----|---------------------------------------|------------------------------------------|
| Purpose | Provides explicit axes objects for fine-grained control and customization | Higher-level interface for quick and simple plotting tasks |
| Creation | Created explicitly using `plt.subplots()` or `plt.axes()` | Axes created automatically by `plt` functions |
| Control | Allows precise control over individual axes properties | Manages figure and axes creation automatically |
| Customization | Enables extensive customization of axes properties (titles, labels, legends, limits, etc.) | Offers a simplified interface for basic plotting tasks |
| Flexibility | Provides more flexibility and advanced features for complex or customized plots | Provides a convenient and concise way to create standard plots |
| Complexity | Requires more code and explicit handling of axes objects | Requires less code and handles figure and axes creation internally |
| Usage | Suitable for complex or customized plots where fine control is needed | Suitable for quick and simple plotting tasks |
| Code Examples | ax.plot(x, y, color='blue') | plt.plot(x, y, color='blue') |
|             | ax.set_xlabel('x label')        | plt.xlabel('x label')                    |
|             | ax.set_ylabel('y label')        | plt.ylabel('y label')                    |
|             | ax.set_title('Title')           | plt.title('Title')                       |
|             | ax.legend()                     | plt.legend()                             |

Remember, both `ax` and `plt` are part of the same Matplotlib library and can be used together. The choice between them depends on the complexity and level of customization required for your specific plotting needs.

## Line Plots

Once we have created an axes, we can use the `ax.plot` function to plot some data.

```python
# By matplotlib

fig = plt.figure(figsize=(5,5), dpi=300)
ax = plt.axes()

x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])
ax.plot(x, y, marker='o')

ax.set_title('Simple Plot')   # Add a title
ax.set_xlabel('x label')      # Add x label
ax.set_ylabel('y label')      # Add y label
```

```python
# For seaborn
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

# Set seaborn style
sns.set(style="whitegrid")

# Create a figure and axes
fig, ax = plt.subplots()

# Plot the line
sns.lineplot(x=x, y=y, ax=ax, color='red', marker='s')

# Add labels and a title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Single Line Plot')
```

Multiple lines on same plot

```python
# By matplotlib

fig = plt.figure()
ax = plt.axes()

x = np.array([1, 2, 3, 4, 5])
y1 = np.array([2, 4, 6, 8, 10])
y2 = np.array([1, 3, 5, 7, 9])

ax.plot(x, y1, marker='o', label='Line 1')
ax.plot(x, y2, marker='s', label='Line 2')

ax.set_title('Multiple Lines Plot')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.legend()  # Add legend to distinguish the lines
```

```python
# For seaborn

x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]

sns.set(style="whitegrid")  # Set seaborn style

# Create a line plot using seaborn
sns.lineplot(x=x, y=y1, marker='o', label='Line 1')
sns.lineplot(x=x, y=y2, marker='s', label='Line 2')

# Set labels and title
plt.title('Multiple Lines Plot')
plt.xlabel('x label')
plt.ylabel('y label')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

Set line colors. There are several ways to specify the color:
- Using a single character color code. For example, 'b' represents blue, 'r' represents red, 'g' represents green, and so on.
- Using a full color name such as 'blue', 'red', 'green', and so on.
- Using a hex color code, such as '#FF0000' for red, '#00FF00' for green, '#0000FF' for blue, and so on.

```python
# For matplotlib

fig = plt.figure()
ax = plt.axes()

x = np.array([1, 2, 3, 4, 5])
y1 = np.array([2, 4, 6, 8, 10])
y2 = np.array([1, 3, 5, 7, 9])

ax.plot(x, y1, color='blue', marker='o', label='Line 1')
ax.plot(x, y2, color='#FF0000', marker='s', label='Line 2')

ax.set_title('Multiple Lines Plot')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.legend()
```

```python
# For seaborn

x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]

# Set seaborn style
sns.set(style="whitegrid")

# Create a line plot using seaborn
sns.lineplot(x=x, y=y1, marker='o', color='blue', label='Line 1')
sns.lineplot(x=x, y=y2, marker='s', color='#FF0000', label='Line 2')

# Set labels and title
plt.title('Multiple Lines Plot')
plt.xlabel('x label')
plt.ylabel('y label')

# Add a legend
plt.legend()
```

Line Styles: you can use the linestyle parameter in the ax.plot() function. The linestyle parameter allows you to specify different line styles for your plot. Here are some commonly used line styles:
- '-' or 'solid': Solid line (default)
- '--' or 'dashed': Dashed line
- '-.' or 'dashdot': Dash-dotted line
- ':' or 'dotted': Dotted line

```python
# For Matplotlib
fig = plt.figure()
ax = plt.axes()

x = np.array([1, 2, 3, 4, 5])
y1 = np.array([2, 4, 6, 8, 10])
y2 = np.array([1, 3, 5, 7, 9])

ax.plot(x, y1, linestyle='-', color='blue', marker='o', label='Line 1')
ax.plot(x, y2, linestyle='--', color='red', marker='s', label='Line 2')

ax.set_title('Multiple Lines Plot')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.legend()
```

```python
# For seaborn
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]

# Set seaborn style
sns.set(style="whitegrid")

# Create a line plot using seaborn
sns.lineplot(x=x, y=y1, linestyle='-', color='blue', marker='o', label='Line 1')
sns.lineplot(x=x, y=y2, linestyle='--', color='red', marker='s', label='Line 2')

# Set labels and title
plt.title('Multiple Lines Plot')
plt.xlabel('x label')
plt.ylabel('y label')

# Add a legend
plt.legend()
```

Enhance the line plots

```python
# For matplotlib

x = [0, 2, 4, 6, 8, 10, 12]
y1 = [0, 0.909, -0.959, 0.986, 0.412, -0.544, -0.999]
y2 = [1, -0.416, -0.654, 0.960, -0.145, -0.911, -0.839]

# Create a figure and axes
fig, ax = plt.subplots()

# Customize Line Style and Color
ax.plot(x, y1, linestyle='--', color='red', label='Line 1')
ax.plot(x, y2, linestyle=':', color='blue', label='Line 2')

# Add Labels and Title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Line Plot')

# Include Legends
ax.legend()

# Adjust Axes Limits
ax.set_xlim(0, 12)
ax.set_ylim(-1, 1)

# Add Gridlines
ax.grid(True)

# Add Annotations
ax.annotate('Point 1', xy=(4, 0.986), xytext=(4, 0.7),
             arrowprops=dict(arrowstyle='->'))
```

```python
# For seaborn

# Sample data
x = [0, 2, 4, 6, 8, 10, 12]
y1 = [0, 0.909, -0.959, 0.986, 0.412, -0.544, -0.999]
y2 = [1, -0.416, -0.654, 0.960, -0.145, -0.911, -0.839]

# Set seaborn style
sns.set(style="whitegrid")

# Create a figure and axes
fig, ax = plt.subplots()

# Plot the lines
sns.lineplot(x=x, y=y1, ax=ax, linestyle='--', color='red', label='Line 1')
sns.lineplot(x=x, y=y2, ax=ax, linestyle=':', color='blue', label='Line 2')

# Add labels and a title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Line Plot')

# Show the legend
ax.legend()

# Adjust the y-axis limits
ax.set_ylim(-1, 1)
```

Load dataset and plot a line plot

```python
# For matplotlib

# Load the iris dataset
iris = sns.load_dataset('iris')
print(iris.head)

# Set seaborn style
sns.set(style="whitegrid")

# Create a line plot of sepal length for each class
for species in iris['species'].unique():
    species_data = iris[iris['species'] == species]
    # print(species_data)
    x = range(len(species_data))
    # print(x)
    y = species_data['sepal_length']
    # print(y)
    plt.plot(x, y, label=species)

# Add labels and a title
plt.xlabel('Index')
plt.ylabel('Sepal Length')
plt.title('Line Plot of Sepal Length by Species')

# Add a legend
plt.legend()
```

```python
# For seaborn

# Load the iris dataset
iris = sns.load_dataset('iris')

# Set seaborn style
sns.set(style="whitegrid")

# Create a line plot of sepal length for each class
sns.lineplot(x=range(len(iris)), y=iris['sepal_length'], hue=iris['species'])

# Add labels and a title
plt.xlabel('Index')
plt.ylabel('Sepal Length')
plt.title('Line Plot of Sepal Length by Species')
```

EDA and visualization with Line plot

```python
# Generate temperature data (dates and corresponding temperatures)
start_date = datetime.date(2023, 11, 1)
end_date = datetime.date(2023, 12, 1)
num_days = (end_date - start_date).days

dates = [start_date + datetime.timedelta(days=i) for i in range(num_days)]
temperatures = np.random.uniform(5, 25, num_days)
print(temperatures)

# Convert dates to Matplotlib-compatible format
x = [datetime.datetime.strptime(str(date), "%Y-%m-%d").date() for date in dates]

# Set Seaborn style
sns.set(style="darkgrid")

# Plotting the line plot using Seaborn
sns.lineplot(x=x, y=temperatures, marker='o', color='blue')

# Customizing the plot
plt.title("Average Daily Temperature (April 2023)")
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.xticks(rotation=45)

# Display the plot
plt.tight_layout()
plt.show()

# Calculate descriptive statistics
mean_temp = np.mean(temperatures)
median_temp = np.median(temperatures)
min_temp = np.min(temperatures)
max_temp = np.max(temperatures)
std_temp = np.std(temperatures)

print("Descriptive Statistics:")
print("Mean Temperature:", mean_temp)
print("Median Temperature:", median_temp)
print("Minimum Temperature:", min_temp)
print("Maximum Temperature:", max_temp)
print("Standard Deviation:", std_temp)
```
