## Bar Plot

A bar plot is a type of chart that presents categorical data with rectangular bars, where the length of each bar corresponds to the value or frequency of a particular category.
-   The x-axis of the bar plot represents the categories or groups being compared, while the y-axis represents the values or frequencies associated with each category.
-   Each bar in the plot represents a specific category, allowing for a visual comparison of the values or frequencies between different categories.

Bar plots can be used to:
-   Compare the values or frequencies of different categories and identify any significant differences or patterns.
-   Display and summarize categorical data in a clear and intuitive manner.
-   Communicate distributions, proportions, or percentages within different categories.

#### Bar plot by Matplotlib

```python
# Choose the height of the bars
height = [10,15,5,20]

# Choose the names of the bars
bars = ['Apple', 'Orange', 'Lemon', 'Lime']

# Create bars
plt.bar(bars, height)

# Add labels to the x-axis and y-axis
plt.xlabel('Bars')
plt.ylabel('Height')
plt.show()
```

```python
# Choose the height of the bars
height = [10,15,5,20]

# Choose the names of the bars
bars = ['Apple', 'Orange', 'Lemon', 'Lime']

# Create names on the x-axis
x_pos = np.arange(len(bars))
print(x_pos)
# Create bars
plt.bar(x_pos, height)
plt.xticks(x_pos, bars)

# Add labels to the x-axis and y-axis
plt.xlabel('Bars')
plt.ylabel('Height')
plt.show()
```

#### Control order
A bar chart is always more insightful when ordered. It allows to understand how items are ranked much quicker.

```python
# By Matplotlib
df = pd.DataFrame ({
        'Group':  ['Apple', 'Orange', 'Lemon', 'Lime'],
        'Value': [10,15,5,20]
})

# Sort the table
df = df.sort_values(by=['Value'],ascending=False)

ax = plt.axes() 

# Create horizontal bars
ax.barh(y=df.Group, width=df.Value, color="blue")

# Add title
ax.set_xlabel('Quantity')
ax.set_ylabel('Fruit')
ax.set_title('Fruit Quantities')
plt.show()
```

```python
# By Seaborn
df = pd.DataFrame({
    'Group': ['Apple', 'Orange', 'Lemon', 'Lime'],
    'Value': [10, 15, 5, 20]
})

# Sort the table
df = df.sort_values(by=['Value'], ascending=True)

# Set the Seaborn style
sns.set(style="whitegrid")

# Create horizontal bar plot using Seaborn
ax = sns.barplot(x='Value', y='Group', data=df, color="blue")

# Set labels and title
ax.set(xlabel='Quantity', ylabel='Fruit', title='Fruit Quantities')
plt.show()
```

## Grouped Bar Plot

```python
#By Matplotlib
labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

x = np.arange(len(labels))  # the label locations
width = 0.25  # the width of the bars

print("x:",x)
print("Men:",x-width/2)
print("Women:",x+width/2)

fig, ax = plt.subplots()
ax.bar(x - width/2, men_means, width=width, label='Men', edgecolor="k")
ax.bar(x + width/2, women_means, width=width, label='Women', edgecolor="k")

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x)
# ax.set_xticklabels(labels)
ax.legend()
plt.show()
```

## Stacked bar plot

```python
labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

width = 0.25  # the width of the bars


fig, ax = plt.subplots()
ax.grid(linestyle='--', color='0.75', axis = 'y')

p1 = ax.bar(labels, men_means, width, label='Men')
p2 = ax.bar(labels, women_means, width, bottom=men_means, label='Women')

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.legend()

# Label with label_type 'center' instead of the default 'edge'
ax.bar_label(p1, label_type='center')
ax.bar_label(p2, label_type='center')
ax.bar_label(p2)
ax.set_ylim(0,70)
```

```python
labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

width = 0.25  # the width of the bars

fig, ax = plt.subplots()
ax.grid(linestyle='--', color='0.75', axis='y')

p1 = ax.bar(labels, men_means, width, label='Men')
p2 = ax.bar(labels, women_means, width, bottom=men_means, label='Women')

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.legend()

# Calculate totals for each group
totals = [men_means[i] + women_means[i] for i in range(len(labels))]

# Label only the totals
ax.bar_label(p2, labels=totals)

ax.set_ylim(0, 70)

plt.show()
```

## Grouped Stacked Bar Plot

```python
def subplot_sentiment(df1, df2, ax):
    colors1 = {'POS': '#4caf50', 'NEU': '#ff9800', 'NEG': '#e51c23'}
    colors2 = {'POS': '#8cbf5b', 'NEU': '#ffd966', 'NEG': '#ff7f7f'}
    barWidth = 0.35
    print(df1.head())
    print(len(df1))
    
    r1 = np.arange(len(df1))
    
    # r1 = np.max(np.concatenate([np.arange(len(df1)), np.arange(len(df2))]))
    r2 = [x + barWidth for x in r1]
    
    ax.set_xlim(r1.min() - barWidth, r1.max() + barWidth*2)
    
    ax.bar(r1, df1['POS'], color=colors1['POS'], width=barWidth, edgecolor='white', label='df1 POS')
    ax.bar(r1, df1['NEU'], bottom=df1['POS'], color=colors1['NEU'], width=barWidth, edgecolor='white', label='df1 NEU')
    ax.bar(r1, df1['NEG'], bottom=df1['POS']+df1['NEU'], color=colors1['NEG'], width=barWidth, edgecolor='white', label='df1 NEG')
    ax.bar(r2, df2['POS'], color=colors2['POS'], width=barWidth, edgecolor='white', label='df2 POS')
    ax.bar(r2, df2['NEU'], bottom=df2['POS'], color=colors2['NEU'], width=barWidth, edgecolor='white', label='df2 NEU')
    ax.bar(r2, df2['NEG'], bottom=df2['POS']+df2['NEU'], color=colors2['NEG'], width=barWidth, edgecolor='white', label='df2 NEG')
    
    ax.set_xticks([r + barWidth/2 for r in range(len(df1))], df1.index)
```

```python
dfT_all = pd.read_excel("sentiment_all.xlsx", sheet_name="TEKKEN7", index_col=0)
dfP_all = pd.read_excel("sentiment_all.xlsx", sheet_name="PUBG", index_col=0)
dfD_all = pd.read_excel("sentiment_all.xlsx", sheet_name="Dota2", index_col=0)
dfC_all = pd.read_excel("sentiment_all.xlsx", sheet_name="CSGO", index_col=0)
    

dfT_less10 = pd.read_excel("sentiment_less_10_hours.xlsx", sheet_name="TEKKEN7", index_col=0)
dfP_less10 = pd.read_excel("sentiment_less_10_hours.xlsx", sheet_name="PUBG", index_col=0)
dfD_less10 = pd.read_excel("sentiment_less_10_hours.xlsx", sheet_name="Dota2", index_col=0)
dfC_less10 = pd.read_excel("sentiment_less_10_hours.xlsx", sheet_name="CSGO", index_col=0)


# Create a figure with four subplots arranged in a vertical layout
fig, axs = plt.subplots(nrows=4, ncols=1, figsize=(24, 14), sharex = True)


# Call the subplot_sentiment function for the first subplot
subplot_sentiment(dfT_all, dfT_less10, axs[0])
subplot_sentiment(dfD_all, dfD_less10, axs[1])
subplot_sentiment(dfP_all, dfP_less10, axs[2])
subplot_sentiment(dfC_all, dfC_less10, axs[3])


labels = ["POS", "NEU", "NEG"]

fig.legend(labels,loc="upper center",fontsize="14",ncol=len(labels),bbox_transform=fig.transFigure,prop={'size': 14})

fig.suptitle(" ", fontsize=16)

axs[0].yaxis.set_major_formatter(mtick.PercentFormatter(1))
axs[1].yaxis.set_major_formatter(mtick.PercentFormatter(1))
axs[2].yaxis.set_major_formatter(mtick.PercentFormatter(1))
axs[3].yaxis.set_major_formatter(mtick.PercentFormatter(1))

axs[0].set_ylabel("TEKKEN7(%)", fontsize=14)
axs[1].set_ylabel("Dota2(%)", fontsize=14)
axs[2].set_ylabel("PUBg(%)", fontsize=14)
axs[3].set_ylabel("CS:GO(%)", fontsize=14)

fig.tight_layout()
```

## Lollipop chart

```python
# Choose the height of the bars
height = [10, 15, 5, 20]

# Choose the names of the bars
bars = ['Apple', 'Orange', 'Lemon', 'Lime']
# Create the Lollipop chart
plt.stem(bars, height)
plt.show()
```

```python
# Load the tips dataset
tips = sns.load_dataset('tips')
print(tips.head())

# Calculate the average total bill for each day
average_total_bill = tips.groupby('day')['total_bill'].mean().reset_index()

# Sort the data by average total bill
average_total_bill = average_total_bill.sort_values(by='total_bill', ascending=False)

# Create the Lollipop chart
sns.set_theme(style="whitegrid")
plt.stem(average_total_bill['total_bill'])

# Set the x-tick labels
plt.xticks(ticks=range(len(average_total_bill)), labels=average_total_bill['day'])

plt.xlabel('Day')
plt.ylabel('Average Total Bill')
plt.title('Average Total Bill by Day')
plt.show()
```
